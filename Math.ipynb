# Plotting Derivatives of f(x) on Python and Mathematica comparison

```python
import sympy
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
x = sympy.Symbol('x')
f = map(sympy.Function, "f")
```

```python
f = x**2 + sympy.sin(x)
lam_f = sympy.lambdify(x,f,modules = ['numpy'])
lam_f_d1 = sympy.lambdify(x, sympy.diff(f), modules = ['numpy'])
lam_f_d2 = sympy.lambdify(x, sympy.diff(sympy.diff(f)), modules = ['numpy'])
```

```python
x_vals = np.linspace(0,5,100)
y_vals = lam_f(x_vals)
y_vals_d1 = lam_f_d1(x_vals)
y_vals_d2 = lam_f_d2(x_vals)
```

```python
plt.title("f(x) = x^2 + sin(x)", fontsize = 15)
plt.plot(x_vals, y_vals, label = "f(x)")
plt.plot(x_vals, y_vals_d1, label = "f'(x)")
plt.plot(x_vals, y_vals_d2, label = "f''(x)")
plt.legend(fontsize = 14)
```
